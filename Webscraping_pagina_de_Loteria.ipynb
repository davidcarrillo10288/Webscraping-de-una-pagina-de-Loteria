```python
## Importamos las librerias que nos servirán para poder hacer scraping a la página de la Lotería
import requests
import lxml.html as html
import os
import datetime
import numpy as np
import pandas as pd
```

```python
## Observamos que la ruta tiene un patrón, donde las páginas terminan en el número correspondiente de la pagina a conseguir los números
ruta = 'https://loteria.guru/resultados-loteria-peru/pe-gana-diario/resultados-anteriores-gana-diario-pe?page=X' ## Observar que le colocamos una "X" al final, para hacer notar que es cambiante este valor
XPATH_NUMEROS_LOTERIA = '//div/ul[@class="lg-numbers-small "]/li[@class="lg-number"]/text()'
XPATH_FECHAS_AÑO = '//div[@class="column is-6 lg-date has-text-right"]/text()[not(@strong)]'
XPATH_FECHAS_MES = '//div[@class="column is-6 lg-date has-text-right"]/strong/text()'
```

```python
## Generamos un array, donde colocaremos las 31 páginas de las cuales haremos scraping (Al dia de hoy)
## Si aumentara el numero de páginas, solo tendríamos que modificar este valor
rutas_totales = []
for i in range(1,32):
  ruta = ruta.replace("X",str(i))
  rutas_totales.append(ruta)
  ruta = ruta.replace(str(i),"X")
print(rutas_totales)

```

```python
TOTAL_NUMBERS = []
TOTAL_FECHAS_MES = []
TOTAL_FECHAS_AÑO = []
JUGADA_POR_FECHA = []
```

 ** CALCULANDO TODOS LOS NUMEROS ALEATORIOS DE LOTERIA, EN PAREJAS DE 5 EN 5**

```python
for ruta in rutas_totales:
  response = requests.get(ruta)
  if response.status_code == 200:
    numbers = response.content.decode('utf-8')
    parsed = html.fromstring(numbers)
    link_to_numbers = parsed.xpath(XPATH_NUMEROS_LOTERIA)
    link_to_fechas_mes = parsed.xpath(XPATH_FECHAS_MES)
    link_to_fechas_año = parsed.xpath(XPATH_FECHAS_AÑO)
    for number in link_to_numbers:
      number = int(number)
      TOTAL_NUMBERS.append(number)
    for mes in link_to_fechas_mes:
      TOTAL_FECHAS_MES.append(mes)
    for año in link_to_fechas_año:
      TOTAL_FECHAS_AÑO.append(año)
    
  else:
    raise ValueError(f'Error: {response.status_code}')
```

```python
#Lo convertimos a array para trabajar mejor
#Este codigo solo me muestra los resultados de la primera pagina, 10 resultados de 5 numeros
TOTAL_NUMBERS
np.array(TOTAL_NUMBERS)
```

```python
len(TOTAL_NUMBERS)
```

```python
## Agrupamos los numeros del array, de 5 en 5, ya que esa es la cantidad de numeros aleatorios por dia
j=0
for i in range(0,int((len(TOTAL_NUMBERS)/5))):
  JUGADA_POR_FECHA.append(TOTAL_NUMBERS[j:j+5])
  j = j+5
print(JUGADA_POR_FECHA)
```

```python
len(JUGADA_POR_FECHA) ## Observamos que tenemos 10 jugadas por página, excepto la última que tiene solo 8 jugadas
## Haciendo calculos, estamos bien en cantidad y datos, verificar de todas maneras
```

```python
## Analizamos las fechas, los arrays de mes y año respectivamente
TOTAL_FECHAS_MES, TOTAL_FECHAS_AÑO
```

```python
## Corrigiendo los valores obtenidos de los años, lo generamos en un array corregido
TOTAL_FECHAS_AÑO_CORREGIDO = [i for i in TOTAL_FECHAS_AÑO if i != '\n        ']
for i in range(len(TOTAL_FECHAS_AÑO_CORREGIDO)):
    TOTAL_FECHAS_AÑO_CORREGIDO[i] = TOTAL_FECHAS_AÑO_CORREGIDO[i].replace(" ","").replace("\n","")

print(TOTAL_FECHAS_AÑO_CORREGIDO)
```

```python
## Creamos un diccionario, donde coloquemos los valores de los tres arrays generados
data_loteria = {'MES':TOTAL_FECHAS_MES,'AÑO':TOTAL_FECHAS_AÑO_CORREGIDO,'JUGADA LOTERIA':JUGADA_POR_FECHA}
```

```python
## Colocamos todos estos valores en un dataframe

df_loteria = pd.DataFrame(data_loteria)
df_loteria
```

PREPARANDONOS PARA ANALIZAR NUESTRAS JUGADAS Y HACER PREDICCIONES

```python
## Nombramos en una nueva variable solo las jugadas obtenidas
df_numeros_prediccion = df_loteria["JUGADA LOTERIA"]
df_numeros_prediccion ## Notamos que cada fila, esta compuesta por una lista, esto debemos descomponerlo
```

```python
len(df_numeros_prediccion) # Numero total de jugadas
```

```python
ACUMULADOR=[] ## Creamos un acumulador, donde colocaremos todas las listas
for i in df_numeros_prediccion:
  ACUMULADOR.append(i)
ACUMULADOR ## Observamos que ahora tenemos una lista, que tiene mas sublistas
```

```python
## Con esto conseguido, simplemente aplicamos la creación del dataframe de pandas
JUGADAS_LOTERIA_SEPARADAS = pd.DataFrame(ACUMULADOR)
JUGADAS_LOTERIA_SEPARADAS
```

```python
## Mejoramos nuestra tabla, para saber que bolilla es cada numero
JUGADAS_LOTERIA_SEPARADAS.columns = ["JUGADA 1","JUGADA 2","JUGADA 3","JUGADA 4","JUGADA 5"]
JUGADAS_LOTERIA_SEPARADAS
```
